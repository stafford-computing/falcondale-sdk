# Advanced example

Well, now that we know how it works we would like to explore a bit more.

```python
from sklearn import datasets

# import some data to play with
X, y = datasets.load_breast_cancer(return_X_y=True, as_frame=True)
X
```

```python
from falcondale import Project

dataset = X.copy()
dataset["target"] = y

myproject = Project(dataset, target="target")
```

```python
myproject.profile_dataset()
```

Done with the loading. Let's reduce the dimensionality of our dataset to train a Quantum Neural Network.

```python
myproject.preprocess()
myproject.show_features()
```

```python
myproject.feature_selection(max_cols=10)
```

This first examples will use default configuration for our Quantum Neural Network.

```python
model = myproject.evaluate("qnn")
```

```python
model.list_metrics()
```

```python
model.metric("accuracy")
```

```python
model.metric("auc")
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X)

print(classification_report(y, y_pred))
```

This very same report over the test sample can be retrieved directly from the model.

```python
model.print_report()
```

We can check how good our classifier is outside the Falcondale SDK as well by retrieving either the prediction or probability for the 1 class so that we can evaluate against a set different than the initially provided one.

```python
from sklearn.metrics import roc_auc_score

y_pred_p = model.predict_proba(X)

print("AUC over the whole dataset ", roc_auc_score(y, y_pred_p))
```

Then, we can add options like the test sample ratio or the layers to be used for our particular QNN architecture. Few options are yet available but the SDK will be opened so that future versions allow for further customization or _auto_ mode for automatic definition of hyperparameters.

```python
model = myproject.evaluate("qnn", test_size=0.1, layers=1)
model.print_report()
```
